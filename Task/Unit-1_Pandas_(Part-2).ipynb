## Topic: Data Exploration
Descriptive statistics provide a numerical summary of the central tendency, dispersion, and shape of a dataset's distribution. This helps in understanding the central values and spread of your data.

**Question 3:** Calculate and display the descriptive statistics (mean, standard deviation, min, max, quartiles) for the 'Score' and 'Attendance' columns in the `df` DataFrame.

```python
# Write your answer here
print("mean of Attendance",df['Attendance'].mean())
print("min of Attendance",df['Attendance'].min())
print("max of Attendance",df['Attendance'].max())
print("standard deviation of Attendance",df['Attendance'].std())
print("quantiles of Attendance(q3)",df['Attendance'].quantile(0.75))
print("quantiles of Attendance(q1)",df['Attendance'].quantile(0.25))

print("mean of Score",df['Score'].mean())
print("min of Score",df['Score'].min())
print("max of Score",df['Score'].max())
print("standard deviation of Score",df['Score'].std())
print("quantiles of Score(q3)",df['Score'].quantile(0.75))
print("quantiles of Score(q1)",df['Score'].quantile(0.25))
```
