```python
# create a newtworkx flow for a sosial group having a 5 people where 
# person 1 follows person 2 and person 3 and person 4
# person 2 follows (1,5)
# person 3 follows (1,2,4,5)
# person 4 follows (3,5)
# person 5 follows (1,4,5)

import networkx as nx

g = nx.DiGraph()
g.add_nodes_from([1,2,3,4,5],)

# g.add_edges_from([(1,2),(1,3),(1,4),(2,1),(2,5),(3,1),(3,2),(3,4),(3,5),(4,3),(4,5),(5,1),(5,4),(5,5)])
l = [[0,1,1,1,0],[1,0,0,0,1],[1,1,0,1,1],[0,0,1,0,1],[1,0,0,1,1]]

for i in range(1,6):
    for j in range(1,6):
        if l[i-1][j-1] == 1:
            g.add_edge(i,j)


nx.draw_networkx(g,edge_color="red")
plt.show()
```

```python
import plotly.express as px
import plotly.io as pio
import pandas as pd

d = {
    "dep":["IT","IT","HR","HR","Finance","Finance"],
    "gender":["M","F","M","F","M","F"],
    "salary":[50000,55000,45000,48000,70000,72000],
    "exp":[3,4,2,3,5,6]
}
df = pd.DataFrame(d)
df
```

```python
# pio.renderers.default = "browser"
pio.renderers.default = "notebook"
```

```python
fig = px.bar(df,"dep","salary",color="gender",barmode="stack",title="abc",height=300,width=500,hover_data=["exp"]
             ,orientation="h",color_discrete_map={"M":"black","F":"yellow"})
fig.show()
```

```python
fig = px.bar(df,"dep","salary",color="gender",barmode="group")
fig.show()
```

```python
fig = px.bar(df,"dep","salary",color="gender",barmode="relative")
fig.show()
```

```python
fig = px.bar(df,"dep","salary",color="gender",barmode="overlay")
fig.show()
```

```python
fig = px.scatter(df,x="salary",y='exp',size="salary",color="gender",symbol="gender")
fig.show()
```

```python
fig = px.scatter_3d(df,x="salary",y='exp',z="dep",size="salary",color="gender")
fig.show()
```

```python
fig = px.scatter(df,x="salary",y='exp',size="salary",color="gender",symbol="gender",facet_row="gender",facet_col="dep")
fig.show()
```

```python
fig = px.line(df,x="salary",y='exp',markers=True)
fig.show()
```

```python
fig = px.line_3d(df,x="salary",y='exp',z="dep",markers=True)
fig.show()
```

```python
fig = px.pie(df,values="salary",names="dep",hover_data="gender"
             ,opacity=0.9,color_discrete_sequence=["red","yellow","blue"])
fig.update_traces(pull=[0.4,0,1])
fig.show()
```

```python
fig = px.histogram(df,x="salary",nbins=4,color="dep",text_auto=True)
fig.show()
```

```python
import dash
from dash import html,dcc
import plotly.express as px
import pandas as pd

data = {
    "player":["Kholi","Rohit","Dhoni"],
    "run":[75,60,50]
}

df = pd.DataFrame(data)
fig = px.bar(df,x='player',y='run')
app = dash.Dash(__name__)
app.layout = html.Div([
    html.H1("Simple Dash app"),
    dcc.Graph(figure=fig)
])

if __name__=="__main__":
    app.run(port=8051)
```
