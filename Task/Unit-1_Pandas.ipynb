## Topic: Cleaning Data
Data cleaning involves handling missing values, correcting errors, and dealing with inconsistencies in your dataset. This is a crucial step before analysis.

**Question 11:** Imagine a new student was added with missing 'ScienceScore' and 'AttendancePercentage'. Create a new DataFrame with this additional row (StudentID: 111, Name: 'Kane', MathScore: 70). How would you represent the missing values in the 'ScienceScore' and 'AttendancePercentage' columns, and what function would you use to check for them?

```python
new_student = {
    'StudentID': 111,
    'Name': 'Kane',
    'MathScore': 70,
    'ScienceScore': float('nan'),
    'AttendancePercentage': float('nan')
}
df_with_missing = pd.concat([df, pd.DataFrame([new_student])], ignore_index=True)
print(df_with_missing.tail())
print(df_with_missing[['ScienceScore', 'AttendancePercentage']].isna().tail())
```

## Topic: Cleaning Data
Missing data can be handled by filling them with a specific value (e.g., mean, median) or by removing rows/columns with missing values.

**Question 12:** For the student DataFrame, assume 'AttendancePercentage' has some missing values represented as NaN. If you wanted to fill these missing values with the average attendance percentage of all students, what Pandas function would you use, and what value would you pass to it?

```python
average_attendance = df_with_missing['AttendancePercentage'].mean()
df_filled = df_with_missing.copy()
df_filled['AttendancePercentage'] = df_filled['AttendancePercentage'].fillna(average_attendance)
print("Average attendance used for fillna:", average_attendance)
print(df_filled.tail())
```

## Topic: Cleaning Data
Data types can sometimes be incorrect, leading to issues in analysis. Converting data types is a common cleaning task.

**Question 13:** If the 'AttendancePercentage' column was initially read as an object (string) type due to a typo like '95.5%', how would you convert it to a numeric (float) type? What function would you use?

```python
df['AttendancePercentage'] = pd.to_numeric(
    df['AttendancePercentage'].astype(str).str.rstrip('%'),
    errors='coerce'
)
print(df['AttendancePercentage'])
print("Converted dtype:", df['AttendancePercentage'].dtype)
```

## Topic: Cleaning Data
Identifying and removing duplicate records is essential to ensure data integrity and prevent biased analysis.

**Question 14:** Suppose the student DataFrame accidentally had a duplicate row for 'Alice' (StudentID 101). How would you identify this duplicate row and then remove it to keep only the first occurrence?

```python
df_with_duplicate = pd.concat([df, df[df['StudentID'] == 101]], ignore_index=True)
duplicate_rows = df_with_duplicate[df_with_duplicate.duplicated(subset=['StudentID', 'Name'], keep=False)]
print("Duplicate rows:")
print(duplicate_rows)
df_no_duplicates = df_with_duplicate.drop_duplicates(subset=['StudentID', 'Name'], keep='first')
print("\nAfter removing duplicates:")
print(df_no_duplicates)
```

## Topic: Cleaning Data
Sometimes, data might be stored in an inconsistent format (e.g., 'USA' vs 'United States'). Standardizing these values is part of cleaning.

**Question 15:** If there was a column 'Grade' in the student DataFrame with values like 'A', 'a', and 'A+', how would you standardize these to a single consistent format like 'A' for analysis?

```python
df_grades = df.copy()
df_grades['Grade'] = ['A', 'a', 'A+', 'B', 'b', 'C', 'A', 'a+', 'B+', 'c']
df_grades['Grade'] = df_grades['Grade'].str.strip().str.upper().replace({'A+': 'A'})
df_grades['Grade'] = df_grades['Grade'].str.strip().str.upper().replace({'B+': 'B'})
print(df_grades[['Name', 'Grade']])
```

## Topic: Data Aggregation & Transformation
Transformation involves changing the structure or values within your data. This can include applying functions to columns or creating new derived columns.

**Question 18:** Create a new column called 'MathGrade' in the student DataFrame. Assign 'A' if 'MathScore' is >= 90, 'B' if between 80 and 89, and 'C' otherwise. Display the DataFrame with this new column.

```python
df['MathGrade'] = 'C'
df.loc[df['MathScore'].between(80, 89), 'MathGrade'] = 'B'
df.loc[df['MathScore'] >= 90, 'MathGrade'] = 'A'
print(df[['Name', 'MathScore', 'MathGrade']])
```

## Topic: Data Aggregation & Transformation
You can apply multiple aggregation functions to different columns simultaneously using the `.agg()` method.

**Question 19:** Using the student DataFrame, calculate the maximum 'ScienceScore' and the minimum 'AttendancePercentage' in a single operation. Display both results.

```python
summary_stats = df[['ScienceScore', 'AttendancePercentage']].agg({
    'ScienceScore': 'max',
    'AttendancePercentage': 'min'
})
print(summary_stats)
```

## Topic: Data Aggregation & Transformation
Reshaping data is a powerful transformation technique used to change the layout of your DataFrame for different analytical needs.

**Question 20:** If you wanted to transform the student DataFrame so that each student has one row and columns for 'MathScore' and 'ScienceScore' are represented as rows under a 'ScoreType' column, what Pandas function would you use?

```python
long_format = pd.concat([
    df[['StudentID', 'Name', 'MathScore']].rename(columns={'MathScore': 'Score'}).assign(ScoreType='MathScore'),
    df[['StudentID', 'Name', 'ScienceScore']].rename(columns={'ScienceScore': 'Score'}).assign(ScoreType='ScienceScore')
], ignore_index=True)[['StudentID', 'Name', 'ScoreType', 'Score']]
print(long_format)
```
