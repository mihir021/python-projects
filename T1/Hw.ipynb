```python
import pandas as pd

df1 = pd.DataFrame(
    {
        'roll':[1,2,3],
        'Name':['A','B','C']
    }
)
df2 = pd.DataFrame(
    {
        'roll':[1,2,4],
        'Marks':[24,20,22]
    }
)
```

```python
pd.merge(df1,df2,on='roll')
```

```python
pd.merge(df1,df2,on='roll',how='outer')
```

```python
pd.merge(df1,df2,on='roll',how='left')
```

```python
pd.merge(df1,df2,on='roll',how='right') # b - a
```

```python
pd.concat((df1,df2))
```

```python
"""
    insert a new col per where if marks if 25 the pre is good 22-25 avg else belove avg
"""

def tracker(marks):
    if marks == 25:
        return "Good"
    elif marks >= 22:
        return "avg"
    else:
        return "belove avg"

df2['per'] = df2['Marks'].apply(tracker)

df2
```
